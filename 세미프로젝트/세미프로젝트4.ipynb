```python
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
userAgent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
```

```python
# 접속객체 생성
session = requests.Session()
# 접속객체에 부가정보(header) 삽입하기
session.headers.update({
    "Referer": "",
    "User-Agent": userAgent
})
```

```python
# 생성한 접속객체를 활용하여 API에 접속
r = session.get(url4)
if r.status_code != 200:
    msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
    raise Exception(msg)
```

```python
# 인코딩 형식 지정하여 beautifulsoup객체를 생성
r.encoding = "euc-kr" # 이 홈페이지는 euc-kr로 해야함
soup = BeautifulSoup(r.text)
soup
```

```python
# soup객체로부터 원하는 부분 추출하기
test4 = soup.select('#articleBody b')
# test4
```

```python
h4 = []
h4_name = []
for item in test4:
    h4 += item.text.strip().split()
# print(h4)
for i in range(1, 7):
    if i%2 == 0:
        h4_name.append(h4[i-1])
for i in range(15, 28):
    if i%2 == 0:
        h4_name.append(h4[i])
for i in range(11, 15):
    if (i+1)%2 == 0:
        h4_name.append(h4[i])
print(h4_name)
```

```python
df = DataFrame(h4_name)
df.to_excel("result4.xlsx")
df
```
