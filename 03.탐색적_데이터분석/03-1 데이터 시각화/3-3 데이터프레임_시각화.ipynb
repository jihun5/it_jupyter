# 데이터프레임 시각화

- 데이터 프리엠이 내부적으로 pyplot 객체를 초기화 하기 때문에 plt.figure() 호출할 필요가 없다.

01. 작업준비

```python
from pandas import read_excel
from matplotlib import pyplot as plt
```

```python
# 기본세팅
# 폰트 설정
#plt.rcParams['font.family'] = "Malgun Gothic" # 윈도우 전용
plt.rcParams['font.family'] = "Malgun Gothic"

#글자 크기
plt.rcParams["font.size"] = 12

# 그래프 크기설정(inch)
plt.rcParams["figure.figsize"] = (10, 5)

# 유티코드에서 음수 부호 설정
plt.rcParams['axes.unicode_minus'] = False
```

02. 데이터가져오기

```python
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx", index_col='년도')
df
```

#02. Line Plot

- 특정 컬럼에 대한 시각화

###### 기본적으로 DataFrame의 컬럼 이름이 label이 되지만 plot() 메서드에 label 파라미터를 설정하여 다른 이름으로 변경할 수 있다.

```python
df['발생건수'].plot(label='교통사고')
plt.grid()
plt.legend()
plt.title("교통사고 발생건수")
plt.show()
plt.close()
```

2. 2개이상 컬럼에 대한 시각화

```python
df.filter(['사망자수', '부상자수']).plot()
plt.grid()
plt.legend()
plt.title("교통사고 사망자수, 부상자수")
plt.show()
plt.close()
```

3. 전체 컬럼에 대한 시각화

```python
df.plot()
plt.grid()
plt.legend()
plt.title("년도별 교통사고")
plt.ylabel("교통사고 수")
plt.show()
plt.close()
```

#03. Bar Plot

- 특정 컬럼에 대한 시각화

```python
df['발생건수'].plot.bar(rot=45) # rot은 아래 글자 기울기
plt.grid()
plt.title("년도별 교통사고 변화")
plt.ylim(200000, 240000)
plt.show()
plt.close()
```

3. 전체 컬럼에 대한 시각화

```python
df.plot.bar(rot=45)
plt.grid()
plt.legend()
plt.title("년도별 교통사고 변화")
plt.show()
plt.close()
```

#04. 산점도 그래프

```python
df.plot.scatter(x='발생건수', y='부상자수')
plt.grid()
plt.show()
plt.close()
```

#05. 파이 그래프

###### dataframe의 index가 파이 그래프의 각 조각이 된다.

```python
df['사망자수'].plot.pie(autopct='%0.1f%%') # 소수점 0.1까지표시
plt.show()
plt.close()
```
