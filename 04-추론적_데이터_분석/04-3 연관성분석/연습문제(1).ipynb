```python
df = read_excel("https://data.hossam.kr/E03/product.xlsx")
df
```

```python
sb.pairplot(df, size=2.5, diag_kind='hist')
plt.show()
```

```python
sb.heatmap(data=df.corr(), square=True, annot=True, cbar=True)
plt.show()
```

```python
r, p = stats.pearsonr(df['x'], df['y'])

"상관계수: %f, p-value: %f" % (r, p)
```

p의 값이 0.05보다 낮기떄문에 x와 y관의 상관관계가 존재하지 않을 것이라는 귀무가설을 기각한다. 따라서 x와 y와의 관계성이 있다.

```python
get_confidence_interval(df, clevel=0.95)
```

```python
sb.kdeplot(data=df, x='x')
sb.lineplot(x=[32, 28], y=[0, 0.1], color='red')
sb.lineplot(x=[39, 33], y=[0, 0.1], color='blue')
plt.show()
plt.close()
```
