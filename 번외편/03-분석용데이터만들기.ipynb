# 분석용 데이터 만들기
#### #01. 작업준비
패키지 가져오기

```python
from pandas import read_excel, merge, DataFrame, get_dummies
```

데이터 불러오기

```python
origin = read_excel('골목상권_코로나전후_매출액비교.xlsx')
origin.head()
```

#### #02. 데이터 가공

지도에 사용할 정보는 제거

```python
df = origin.drop(['기준_년_코드','기준_분기_코드', '엑스좌표_값','와이좌표_값','시군구_코드','행정동_코드','시군구명','시군구코드'], axis=1)
df.head()
```

코로나 전후로 데이터 분리

```python
before = df[df['Covid19'] == 0]
before.head()
```

```python
after = df[df['Covid19'] == 1]
after.head()
```

상권별로 집계

```python
before_df = before.groupby(['상권_코드','상권_코드_명']).mean()
before_df.head()
```

```python
after_df = after.groupby(['상권_코드','상권_코드_명']).mean()
after_df.head()
```

```python
mdf = merge(before_df, after_df, left_index=True, right_index=True)
mdf.head()
```

종속변수 만들기

```python
mdf['매출증감'] = (mdf['분기당_매출_금액_y'] - mdf['분기당_매출_금액_x']) > 0
mdf['매출증감'] = mdf['매출증감'].astype(int)
mdf.head()
```

학습을 위한 최종 데이터 선정

```python
print(mdf.columns)
```

```python
tdf = mdf[['점포수_x', '도심_x', '총_직장_인구_수_x', '총_생활인구_수_x', '점포수_y', '총_직장_인구_수_y', '총_생활인구_수_y', '매출증감']]

tdf.to_excel("tdf.xlsx")
tdf.head()
```
