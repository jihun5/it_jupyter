```python
df = pd.read_excel('jobs_by_scale.xlsx')
df
df1 = list(df['종사자규모별'])
df2 = list(df[2015])
df3 = list(df[2016])
plt.figure()
plt.title("2015-2010")
# 각 막대의 색상을 단색으로 지정
plt.bar(df1, df2, color='#0066ff')
plt.grid()
plt.legend()
plt.show()
plt.close()

plt.figure()
plt.title("2015-2010")
# 각 막대의 색상을 단색으로 지정
plt.bar(df1, df3, color='#0066ff')
plt.grid()
plt.legend()
plt.show()
plt.close()
sns.pairplot(vars=["종사자규모별", 2015, 2016], data=df)
plt.show()
```
