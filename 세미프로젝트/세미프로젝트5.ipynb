```python
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
# 관심사와 관련된 데이터를 수집할 수 있는 웹사이트
url1 = 'https://brunch.co.kr/@gorrajeju/775'
url2 = 'https://menupick.tistory.com/15'
url3 = 'https://welcome8007.com/33'
url4 = 'https://www.maybugs.com/news/articleView.html?idxno=581068'
url5 = 'https://www.daily.co.kr/life312718976#lifeback'
```

```python
userAgent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
```

```python
# 접속객체 생성
session = requests.Session()
# 접속객체에 부가정보(header) 삽입하기
session.headers.update({
    "Referer": "",
    "User-Agent": userAgent
})
```

```python
# 생성한 접속객체를 활용하여 API에 접속
r = session.get(url5)
if r.status_code != 200:
    msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
    raise Exception(msg)
```

```python
# 인코딩 형식 지정하여 beautifulsoup객체를 생성
r.encoding = "euc-kr" # 이 홈페이지는 euc-kr로 해야함
soup = BeautifulSoup(r.text)
# soup
```

```python
# soup객체로부터 원하는 부분 추출하기
test5 = soup.select('strong')
test5
```

```python
h5 = []
h5_name = []
for item in test5:
    h5 += item.text.strip().split()
# print(h5)
for i in range(1, 11):
    if i%2 == 0:
        h5_name.append(h5[i])
for i in range(12, 22):
    if (i+1)%2 == 0:
        h5_name.append(h5[i])
print(h5_name)
```

```python
df = DataFrame(h5_name)
df.to_excel("resultr.xlsx")
df
```
