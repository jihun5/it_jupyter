#데이터 전처리 (1) 행, 열 재배치

- 단일 데이터 프레임에 대한 작업
- 행 혹은 열(변수)에 대한 재배치, 이름 변경
- 정렬
- 특정 데이터 필터링
- 행 혹은 열 추가, 삭제

```python
from pandas import read_excel
from pandas import Series
import numpy
```

# 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/pre_sample.xlsx", index_col="이름")
df
```

# 데이터 확인하기

```python
df.shape
```

```python
df.info()
```

# 데이터 타입 변경

```python
df2 = df.astype({'국어': 'float', '영어': 'float', '수학': 'str', '과학': 'bool'})
df2
```

```python
df2.info()
```

# 행 순서 변경
- 지정된  순서대로 행이 재정렬 된 결과가 반환된다.
- 원본은 변화 없음, 결과가 적용된 복사본이 생성된다.
- 재배치 되지 못한 행은 제외된다.

```python
df2 = df.reindex(index=['석영', '호영', '민철', '수현'])
df2
```

# 행, 열의 이름 변경
- 기존이름 : 새이름 형식의 딕셔너리로 지정
- colums와 index 중 변경을 원하는 하나만 설정가능
- 원본은 변화 없음, 결과가 적용된 복사본이 생성된다.
- 이름이 변경되지 않은 항목이 결과에서 제외되지는 않는다
- 원본에 즉시 반영해야 할 경우 inplace=True 파라미터를 적용

```python
df3 = df.rename(columns={'국어': 'kor', '영어': 'eng', '수학': 'math', '과학': 'sci'},
                 index={'석영': 'suk', '호영': 'ho', '민철': 'min', '수현': 'su'}
                 )
df3
```

# 정렬
특정 변수(열, 컬럼)를 기준으로 하여 행 단위로 정렬

- 변수에 대한 오름차순 정렬

- 원본에 즉시 반영해야 할 경우 inplace=True 파라미터를 적용

```python
df4 = df.sort_values('국어')
df4
```

복수 칼럼에 대한 정렬

- 정렬 기준이 되는 컬럼 이름을 리스트로 설정하면 첫 번째 컬럼의 값이 동일한 경우 두번째 칼럼을 기준으로 정렬한다.

```python
df5 = df.sort_values(['영어', '학년'])
df5
```

변수에 대한 내림차순

- ascending=False 파라미터를 추가한다.(기본값=True,오름차순)

```python
df6 = df.sort_values('국어', ascending=False)
df6
```

인덱스를 기준으로 하는 정렬

- 인덱스는 하나만 존재하므로 열 이름을 지정할 필요가 없다
- 정렬 방향을 위한 ascending 파라미터를 False로 설정하면 내림차순으로 정렬된다.
- inplace=True를 적용할 경우 원본에 즉시 반영된다.

```python
df7 = df.sort_index() # 이름 가나다 순으로 정렬되었네
df7
```

특정 데이터 필터링

- 행 단위 필터링(조건 검색)
- 특정 조건에 맞는 행 조회

```python
r1 = df.query('국어 > 80')
r1
```

AND 조건

```python
r2 = df.query('국어 > 80 and 수학 > 80')
r2
```

OR 조건

```python
r3 = df.query('국어 > 80 or 수학 > 80')
r3
```

조건 값을 변수로 적용하기

```python
point = int(input('점수를 입력하세요 :'))
print("입력한 점수인 %d점 보다 높은 학생은 다음과 같습니다." % point)

r4 = df.query('국어 > @point')
r4
```

열 단위 필터링

```python
fl = df.filter(['국어', '영어'])
fl
```

행 혹은 열 추가, 삭제

- 행추가 (원본 자체가 수정됨)

- 리스트로 추가하기
- 추가될 행의 인덱스 이름을 지정해야 한다.
- 리스트로 추가할 경우 dataframe의 컬럼 순서에 맞게 지정해야한다.
- 누락되는 값이 있거나 값의 수가 초과할 경우 에러

```python
df.loc['정호'] = [2, '남자', 90,80,70,60]
df
```

```python
df.loc['민정'] = {'성별': '여자', '학년': 4, '국어': 81, '영어': 72, '수학': 84, '과학': 90}
df
```

시리즈 객체로 추가하기

- 시리즈 객체의 index가 지정되지 않은 경우 값을 추가할 수 없다

```python
s = Series([2, '남자', 90,80,70,60], index=['학년', '성별', '국어', '영어', '수학', '과학'])
df.loc['형석'] = s
df
```

빈값이 있는 경우

- Series 객체 형태만 가능

```python
s = Series([2, '남자', 93, 71, 68], index=['학년', '성별', '국어', '수학', '과학'])
df.loc['석호'] = s
df
```

기존의 행 복사하기

```python
df.loc['철민'] = df.loc['철수']
df
```

행 삭제

- 단일 행 삭제
- 추가는 원본 자체가 변하지만 삭제는 결과가 반영된 새로운 복사본이 생성된다.
- drop() 매서드도 inplace=True 파라미터 적용가능

```python
k1 = df.drop('철민')
k1
```

다중 행 삭제

```python
k2 = df.drop(['석영', '정호', '민정', '형석'])
k2
```

index에 대한 인덱싱으로 삭제하기

```python
k3 = df.drop(df.index[0])
k3
```

index에 대한 슬라이싱으로 삭제하기

```python
k4 = df.drop(df.index[0:10])
k4
```

열 추가 (k4에서 이어서 진행)

- 리스트로 추가하기

```python
k4['한국사'] = [72,56,90,88]
k4
```

딕셔너리로 추가하기

```python
k4['세계사'] = {"민정": 90, "형석": 88, "석호": 72, "철민": 56}
k4
```

Series 객체로 추가하기

- 각 값이 연결될 행의 이름(index)를 지정해야 한다.
- 부분적으로 값을 비워 둘 수 있다.

```python
s = Series([72, 56, 90, 88], index=['석호', '철민', '민정', '형석'])
k4['윤리'] = s
k4
```

단일 값 추가하기

```python
k4['체육'] = 100
k4
```
