# 행 순서 변경
- 지정된  순서대로 행이 재정렬 된 결과가 반환된다.
- 원본은 변화 없음, 결과가 적용된 복사본이 생성된다.
- 재배치 되지 못한 행은 제외된다.

```python
df2 = df.reindex(index=['석영', '호영', '민철', '수현'])
df2
```

# 행, 열의 이름 변경
- 기존이름 : 새이름 형식의 딕셔너리로 지정
- colums와 index 중 변경을 원하는 하나만 설정가능
- 원본은 변화 없음, 결과가 적용된 복사본이 생성된다.
- 이름이 변경되지 않은 항목이 결과에서 제외되지는 않는다
- 원본에 즉시 반영해야 할 경우 inplace=True 파라미터를 적용

```python
df3 = df.rename(columns={'국어': 'kor', '영어': 'eng', '수학': 'math', '과학': 'sci'},
                 index={'석영': 'suk', '호영': 'ho', '민철': 'min', '수현': 'su'}
                 )
df3
```

# 정렬
특정 변수(열, 컬럼)를 기준으로 하여 행 단위로 정렬

- 변수에 대한 오름차순 정렬

- 원본에 즉시 반영해야 할 경우 inplace=True 파라미터를 적용

```python
df4 = df.sort_values('국어')
df4
```

복수 칼럼에 대한 정렬

- 정렬 기준이 되는 컬럼 이름을 리스트로 설정하면 첫 번째 컬럼의 값이 동일한 경우 두번째 칼럼을 기준으로 정렬한다.

```python
df5 = df.sort_values(['영어', '학년'])
df5
```

변수에 대한 내림차순

- ascending=False 파라미터를 추가한다.(기본값=True,오름차순)

```python
df6 = df.sort_values('국어', ascending=False)
df6
```

인덱스를 기준으로 하는 정렬

- 인덱스는 하나만 존재하므로 열 이름을 지정할 필요가 없다
- 정렬 방향을 위한 ascending 파라미터를 False로 설정하면 내림차순으로 정렬된다.
- inplace=True를 적용할 경우 원본에 즉시 반영된다.

```python
df7 = df.sort_index() # 이름 가나다 순으로 정렬되었네
df7
```

조건별 파생변수

```python
df5 = df.copy()
df5
```

```python
df5['합계'] = df.loc[:, '국어':'과학'].sum(axis=1)
df5['평균'] = df.loc[:, '국어':'과학'].mean(axis=1)
df5
```

```python
# 평균에 따른 조건부 파생변수
df5['결과'] = numpy.where(df5['평균'] >=80, '합격', '불합격')
# df5
```

```python
# 경우에 수에 따른 조건 값
conditions = [ (df5['평균']>=90), #'A',
              (df5['평균']>=80), #'B',
              (df5['평균']>=70), #'C' 
             ]
value = ["A", "B", "C"]
df5['학점'] = numpy.select(conditions, value, default='F')
df5
```
