```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', ' 컴퓨터', '선풍기','선풍기','믹서','믹서']

encoder = LabelEncoder()

encoder.fit(items)

labels = encoder.transform(items)
print("인코딩 변환값:", labels)
```

```python
print('인코딩 클래스:', encoder.classes_)
```

```python
print('디코딩 원본 값:', encoder.inverse_transform([1, 2, 5, 0, 4, 4, 3, 3]))
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items = ['TV', '냉장고', '전자렌지', '컴퓨터', '선풍기','선풍기','믹서','믹서']

encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
labels
```

```python
# 두번째, 2차원 데이터로 변환
labels = labels.reshape(-1,1)
labels
```

```python
# 마지막으로 원-핫 인코딩 적용
oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)

print(oh_labels.shape)
oh_labels.toarray()
# to_df = pd.DataFrame(oh_labels)
# to_df.head()
```

```python
df = pd.read_csv('abalone.csv')
df.head()
```

```python
df.columns = ['sex','length','diameter','height', 
                  'whole_weight','shucked_weight','viscera_weight', 
                  'shell_weight','rings']
df
```

```python
df.shape
```

```python
# 데이터 결측값 확인
df.isnull().sum()
```

```python
df.describe()
```

```python
# 전복 성별(m, f ,i)에 따라 groupby 함수를 통해 집계

grouped = df['whole_weight'].groupby(df['sex'])
grouped
```

```python
grouped.sum()
```

```python
# 그룹변수가 하나가 아닌, 전체 연속성 변수에 대한 집계
df.groupby(df['sex']).mean()
```

```python
df.groupby('sex').mean()
```

```python
df['length_bool'] = np.where(df['length']> df['length'].mean(),
                            'length_long',
                            'length_short'
)
df['length_bool']
```

```python
# 그룹변수를 2개이상 선택하여 총계정리
df.groupby(['sex', 'length_bool']).mean()
```

```python
# 더 간결하게 표현한다면
## 그룹변수('sex', 'length_bool') 선택 및 집계변수
df.groupby(['sex', 'length_bool'])['whole_weight'].mean()
```
