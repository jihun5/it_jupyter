```python
from matplotlib import pyplot as plt
from pandas import read_excel
import seaborn as sb
import datetime as dt
import numpy as np
```

```python
df = read_excel("https://data.hossam.kr/D01/mpg.xlsx")
df
```

```python
sb.catplot(data=df,
            x='manufacturer',     # 집단을 구분하는 기준
            kind='count',    # 빈도 막대그래프 그리기 
        )
plt.show()
```

```python
sb.catplot(data=df,
            x='manufacturer',     
            kind='count',
            hue='year'    
        )
plt.show()
```

```python
sb.catplot(data=df,
            x='drv',     
            kind='count',    
        )
plt.show()
```

```python
sb.catplot(data=df,
            x='drv',     
            kind='count',
            hue='cty'    
        )
plt.show()
```
