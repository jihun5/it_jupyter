```python
from sklearn.preprocessing import LabelEncoder
items = ['TV', '냉장고', '전자레인지', '컴퓨터', '선풍기','선풍기','믹서','믹서']
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
print("인코딩 변환값 :", labels)
```

```python
print('인코딩 클래스:', encoder.classes_)
```

```python
print('디코딩 원본 값:', encoder.inverse_transform([0, 1, 4, 5, 3, 3, 2, 2]))
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
items = ['TV', '냉장고', '전자레인지', '컴퓨터', '선풍기','선풍기','믹서','믹서']
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
labels
```

```python
# 2차원 데이터로 변한
labels = labels.reshape(-1,1)
labels
```

```python
# 마지막으로 원-핫 인코딩 적용
oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)

print(oh_labels.shape)
oh_labels.toarray()
```

```python
# 전복 성별(m,f,i)에 따라 groupby 함수를 통해 집게를 할 예정
# 그룹별 합계 : grouped.sum()
# 그룹별 평균 : grouped.mean()
# 그룹별 크기 : grouped.size()
# ex) DataFrame[집계변수].groupby(dataframe[집계대상])
grouped = df['whole_weight'].groupby(df['sex'])
grouped
```

```python
# 그룹변수가 하나가 아닌, 전체 연속형 변수에 대한 집계
df.groupby(df['sex']).describe()
```

```python
df.groupby('sex').mean()
```

```python
df.groupby(df['sex']).max()
```

```python
# 조건에 맞는 변수 추가하기
df['length_bool'] = np.where(df['length']>df['length'].median(),
                             'length_long', #참일경우
                             'length_short' # False 일경우
                             )
df['length_bool']
```

```python
# 그룹변수를 2개 이상 선택하여 총계정리
df.groupby(['sex', 'length_bool']).mean()
```

```python
# 더 간결하게 표현하면
# 그룹변수('sex', 'length_bool') 선택 및 집계변수('whole_weight')의 평균
df.groupby(['sex', 'length_bool'])['whole_weight'].mean()
```

apply 함수 사용

dataframe 타입의 객체에서 호출가능한 apply 함수에 대해 알아보자 

본인이 원하는 행과 열에 연산 혹은 function을 적용할 수 있다.

열 기준으로 집계하고 싶은 경우 axis=1

행 기준으로 집계하고 싶은 경우 axis=0

```python
df[['diameter']].apply(np.average, axis=0)
```

```python
# 행 기준 집계
df[['diameter', 'whole_weight']].apply(np.average, axis=0)
```

```python
# 열 기준 집계
df[['diameter', 'whole_weight']].apply(np.average, axis=1)
```

```python
# 사용자 함수를 통한 집계
import math
def avg_ceil(x,y,z):
    return math.ceil((x+y+z)/3)
df[['diameter', 'height', 'whole_weight']].apply(lambda x : avg_ceil(x[0], x[1], x[2]), axis=1)
```
