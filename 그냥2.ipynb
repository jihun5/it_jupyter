```python
from sklearn.preprocessing import LabelEncoder

itmes = ['TV', '냉장고', '전자렌지', '컴퓨터', '선풍기', '선풍기', '믹서', '믹서']

# LavelEncoder 클래스를 encoder 객체로 생성한 후
encoder = LabelEncoder()

encoder.fit(itmes)

labels = encoder.transform(itmes)
print("인코딩 변환값", labels)
```

```python
print("인코딩 클래스:", encoder.classes_)
```

```python
print('디코딩 원본 값:', encoder.inverse_transform([0, 1, 4, 5, 3, 3, 2, 2]))
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items = ['TV', '냉장고', '전자렌지', '컴퓨터', '선풍기','선풍기','믹서','믹서']
```

```python
# 먼저 숫자 값으로 변환하기 위해 LabelEncoder로 변환
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
labels
```

```python
# 두번째, 2차원 데이터로 변환
labels = labels.reshape(-1,1)
labels
```

```python
oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)

print(oh_labels.shape)
oh_labels.toarray()
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
```

```python
print('feature 들의 평균 값')
print(iris_df.mean())
print("\n feature 들의 분산 값")
print(iris_df.var())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
iris_scaled
```

```python
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print("feature들의 평균 값")
print(iris_df_scaled.mean())
print("\n feature 들의 분산 값")
print(iris_df_scaled.var())
iris_df_scaled
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
```

```python
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print("feature들의 최소 값")
print(iris_df_scaled.min())
print("\n featrue들의 최대값")
print(iris_df_scaled.max())
```
