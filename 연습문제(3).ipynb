```python
from matplotlib import pyplot as plt
from pandas import read_excel
import seaborn as sb
import datetime as dt
import numpy as np
```

```python
# 폰트 설정
#plt.rcParams['font.family'] = "Malgun Gothic" # 윈도우 전용
plt.rcParams['font.family'] = "Malgun Gothic"
#글자 크기
plt.rcParams["font.size"] = 12
# 그래프 크기설정(inch)
plt.rcParams["figure.figsize"] = (15, 7)
# 유티코드에서 음수 부호 설정
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = read_excel("https://data.hossam.kr/D01/mpg.xlsx")
df
```

```python
sb.catplot(data=df,
            x='manufacturer',     # 집단을 구분하는 기준
            kind='count',    # 빈도 막대그래프 그리기 
        )
plt.show()
```

```python
sb.catplot(data=df,
            x='manufacturer',     
            kind='count',
            hue='year'    
        )
plt.show()
```

```python
sb.catplot(data=df,
            x='drv',     
            kind='count',    
        )
plt.show()
```

```python
sb.catplot(data=df,
            x='drv',     
            kind='count',
            hue='cty'    
        )
plt.show()
```

```python
fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()

sb.lineplot(data=df, x=df['manufacturer'], y='cty', color='#ff6600', ax=ax1)
sb.lineplot(data=df, x=df['manufacturer'], y='hwy', color='#0066ff', ax=ax2)

plt.title('자동차제조사 별 ')   
plt.xlabel("cty")      
plt.ylabel("hwy")   
plt.grid()
plt.show()               
plt.close()              
```

```python
for col in ['cty', 'hwy']:
    sb.displot(x=col, hue='displ', kind='hist', data=df)
plt.show()
```

```python
sb.pairplot(df[["displ", "cty", "hwy"]])
plt.show()
```

```python
기울기, 절편 = np.polyfit(df['displ'], df['cty'], 1)
z = np.polyfit(df['displ'], df['cty'], 1)
print("기울기(가중치) = %f" % 기울기)
print("절편(편향) = %f" % 절편)
# np.polyfit()
f = np.poly1d(z)

sb.scatterplot(data=df, x='displ', y='cty')
sb.scatterplot(data=df, x='displ', y='hwy', color='red')
plt.plot(df['displ'], f(df['displ']))
plt.plot()
plt.show()
```
